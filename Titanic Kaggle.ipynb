```python
def predict(data):
    res = pd.DataFrame()
    res['PassengerId'] = data['PassengerId']
    res['Survived'] = (data['Sex'] == 'female').apply(int);
    #res['Survived2'] = res['Survived'].apply(int)
    return res
```

```python
test_data = pd.read_csv('test.csv')
output = predict(test_data)
output
```
